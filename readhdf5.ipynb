```python
import numpy as np
import h5py
```

```python
f = h5py.File('snapshot_001.hdf5', 'r')# read .hdf5 file
f.keys() 
```

```python
data=f['PartType1']
data.keys()
```

```python
pos=data['Coordinates'] # position information of dark matter particles
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection ='3d')
x = pos[:,0]
y = pos[:,1]
z = pos[:,2]
ax.scatter(x, y, z, c = 'r', s=0.0001)
plt.show()# 3D structure of dark matter
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6, 6))
x = pos[:,0]
y = pos[:,1]
z = pos[:,2]
plt.scatter(x,y,s=0.0001)
plt.show() # 2D structure of dark matter
```
